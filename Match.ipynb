```python
import numpy as np
import pandas as pd
```

```python
bk = pd.read_csv('./data2/booking_servicios_sumas_clasificados.csv')
ta = pd.read_csv('./data2/trip_servicios_sumas_clasificados.csv')
```

```python
bk.name = bk.name.str.replace("OYO ", "").str.replace("Cabanas","Cabañas").str.replace("Cabana","Cabañ1") #OYO: On Your Own
bk
```

```python
ta.name = ta.name.str.replace(" Alojamiento","").str.replace(" Hospedaje","").str.replace("OYO ", "").str.replace("Cabanas","Cabañas").str.replace("Cabana","Cabañ1")
ta
```

```python
bk_names = np.loadtxt('./data2/hotelesBK.txt', dtype=str, delimiter="\n")
ta_names = np.loadtxt('./data2/hotelesTA.txt', dtype=str, delimiter="\n")
```

```python
ta = ta.replace({"name":{t:b for t,b in zip(ta_names,bk_names)}})
```

```python
bk[bk.name.isin(ta.name)]
```

```python
shared = pd.merge(left=ta, right=bk, on="name", how="inner", suffixes=("_ta","_bk"))
shared
```

```python
shared = shared[shared.city_bk==shared.city_ta]
shared
```

```python
shared.drop(["id_ta","id_bk","city_bk",], axis=1, inplace=True)
shared = shared.rename(columns={"city_ta":"city"})
```

```python
shared.to_csv('./data2/hoteles_compartidos.csv', index=False)
```
