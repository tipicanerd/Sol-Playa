```python
import numpy as np
import pandas as pd
```

# Booking

```python
booking = pd.read_csv('./booking/bookingSolPlaya.csv', decimal=',')
booking
```

```python
booking.info()
```

```python
booking.drop(['airports', 'beaches','popular_services','url'],axis=1,inplace=True)
booking = booking.drop_duplicates()
```

## City

```python
booking.city = booking.city.apply(lambda x: str(x).replace(",","\n").strip().split("\n")[-1])
booking
```

```python
booking.city = booking.city.str.replace('Centro Tulum','Tulum')
booking
```

```python
booking.sort_values('city',inplace=True)
```

```python
booking['id'] = np.arange(booking.shape[0]) + 1
```

```python
booking['rank'] = np.int64(booking.groupby('city')['id'].rank('first'))
```

```python
booking
```

```python
booking.dropna(subset=["name","rating"],inplace=True)
```

## Comments

```python
booking.comments = booking.comments.apply(lambda x: x.split()[0] if str(x)!="nan" else 0)
booking
```

## Name

```python
booking.name = booking.name.apply(lambda x: x.strip())
booking
```

## Prices

```python
booking.prices = booking.prices.apply(lambda x: x.replace("MXN","").replace(",","").replace("\n\n","\n").strip().replace("\n"," #"))
booking
```

## Services

```python
booking.services = booking.services.apply(lambda x: " # ".join(x.replace(",","").strip().replace("\n",";").split(";;;;;;;;;")).replace(";",""))
booking
```

```python
PM = [
    'Compostela', 'Isla Aguada', 'Bacalar', 'Isla Mujeres', 
    'Sayulita', 'Todos Santos', 'Mazunte', 'Loreto', 
    'Tulum', 'Centro Tulum', 'Sisal', 'Papantla de Olarte'
]
```

```python
booking = booking[booking.city.isin(PM)]
booking
```

```python
booking.city.unique()
```

```python
booking = booking.drop_duplicates()
```

```python
booking.to_csv('./data2/booking_clean.csv',index=False)
```
